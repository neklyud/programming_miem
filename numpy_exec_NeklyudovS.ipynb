```python
#44
rndm = np.random.seed(1234)    # For reproducibility, consider seeding your generator
coord = np.random.normal(size=(10,2), scale = 9)
print(coord)
```

```python
print((coord[:, 1] > 0))
print((coord[:, 0] > 0))
```

```python
y = coord[:, 1]
x = coord[:, 0]
```

```python
phi = np.zeros(len(coord))
first_quad = np.where((x > 0) == (y >= 0))
phi[first_quad] = np.arctan(y[first_quad]/x[first_quad])
second_quad = np.where((x > 0) == (y < 0))
phi[second_quad] = np.arctan(y[second_quad]/x[second_quad]) + 2*np.pi
print(phi)
r = np.sqrt(np.power(coord[:,0], 2) + np.power(coord[:,1], 2))
print(r)
```

```python
x = r * np.cos(phi)
y = r * np.sin(phi)
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
